# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.acik = False
        self.ses = 50
        self.kanallar = ["TRT", "CNN", "FOX", "NTV", "Bloomberg"]
        self.anlik_kanal = 0

    def tv_ac(self):
        if self.acik:
            print("Televizyon zaten açık!")
        else:
            self.acik = True
            print("Televizyon açıldı.")

    def tv_kapat(self):
        if not self.acik:
            print("Televizyon zaten kapalı!")
        else:
            self.acik = False
            print("Televizyon kapatıldı.")

    def ses_ayarla(self):
        yeni_ses = input("Ses seviyesini giriniz (0-100): ")
        yeni_ses = int(yeni_ses)
        if yeni_ses < 0 or yeni_ses > 100:
            print("Geçersiz ses seviyesi!")
        else:
            self.ses = yeni_ses
            print("Ses seviyesi", self.ses)

    def kanal_ekle(self):
        kanal_adi = input("Kanal adını giriniz: ")
        self.kanallar.append(kanal_adi)
        print("Kanal eklendi.")

    def kanal_sayisi(self):
        print("Toplam", len(self.kanallar), "kanal var.")

    def rastgele_kanala_gecis(self):
        self.anlik_kanal = random.randint(0, len(self.kanallar)-1)
        print("Şu anki kanal:", self.kanallar[self.anlik_kanal])

    def tv_bilgileri(self):
        print("Televizyon Durumu:", "Açık" if self.acik else "Kapalı")
        print("Ses Seviyesi:", self.ses)
        print("Kanal Listesi:")
        for kanal in self.kanallar:
            print("-", kanal)
```

```python
k = Kumanda()

while True:
    print("**********************")
    print("Kumanda Menüsü")
    print("1. TV Aç")
    print("2. TV Kapat")
    print("3. Ses Ayarları")
    print("4. Kanal Ekle")
    print("5. Kanal Sayısı")
    print("6. Rastgele Kanala Geç")
    print("7. Televizyon Bilgileri")
    print("Çıkmak için q tuşuna basın.")
    print("**********************")

    secim = input("İşlem seçiniz: ")

    if secim == "q":
        print("Programdan çıkılıyor...")
        break

    elif secim == "1":
        k.tv_ac()

    elif secim == "2":
        k.tv_kapat()

    elif secim == "3":
        k.ses_ayarla()

    elif secim == "4":
        k.kanal_ekle()

    elif secim == "5":
        k.kanal_sayisi()

    elif secim == "6":
        k.rastgele_kanala_gecis()

    elif secim == "7":
        k.tv_bilgileri
```
