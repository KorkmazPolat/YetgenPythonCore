# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, capacity):
        self.capacity = capacity

    def seating_capacity(self):
        return f"The seating capacity of this vehicle is {self.capacity} passengers."


class Bus(Vehicle):
    def __init__(self, capacity=50):
        Vehicle.__init__(self, capacity)

````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
class Vehicle:
    def __init__(self, capacity):
        self.capacity = capacity

    def seating_capacity(self):
        return f"The seating capacity of this vehicle is {self.capacity} passengers."


class Bus(Vehicle):
    def __init__(self, capacity=50):
        Vehicle.__init__(self, capacity)


School_bus = Bus(40)
print(isinstance(School_bus, Vehicle))  # prints "True"
print(School_bus.seating_capacity())  # prints "The seating capacity of this vehicle is 40 passengers."
````
